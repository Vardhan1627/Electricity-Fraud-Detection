```python
#### NOW WE ARE GOING TO TRAIN MODEL BY MACHINE LEARNING
# Used for data handling
import pandas as pd
import numpy as np

# Used to split data into training and testing sets
from sklearn.model_selection import train_test_split

# Machine Learning models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Evaluation metrics
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    average_precision_score,
    confusion_matrix,
    classification_report
)

# Used to standardize features for Logistic Regression
from sklearn.preprocessing import StandardScaler

# Used to create a pipeline
from sklearn.pipeline import Pipeline
```

```python
# X = input features used by the ML model
X = model_df[feature_columns]

# y = target/output column
# FLAG:
# 0 = Normal Consumer
# 1 = Potential Fraud Consumer
y = model_df["FLAG"]
```

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,                         # Features
    y,                         # Target

    test_size=0.20,            # 20% testing data

    random_state=42,           # Same result every time

    stratify=y                 # Maintains 92.65% / 7.35% ratio
)
```

```python
print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

**Model 1: Logistic Regression**

```python

# Create a pipeline:
# 1. Scale the features
# 2. Train Logistic Regression

logistic_model = Pipeline([

    # Standardize feature values
    ("scaler", StandardScaler()),

    # Train Logistic Regression
    # class_weight="balanced" helps with imbalanced data
    ("model", LogisticRegression(
        class_weight="balanced",
        max_iter=1000,
        random_state=42
    ))
])
```

```python
# Train the model using training data
logistic_model.fit(X_train, y_train)
```

```python
# Predict class labels
y_pred_logistic = logistic_model.predict(X_test)

# Predict probability of fraud
y_prob_logistic = logistic_model.predict_proba(X_test)[:, 1]
```

```python
print("LOGISTIC REGRESSION RESULTS")
print("--------------------------------")

print(
    "Accuracy:",
    accuracy_score(y_test, y_pred_logistic)
)

print(
    "Precision:",
    precision_score(y_test, y_pred_logistic)
)

print(
    "Recall:",
    recall_score(y_test, y_pred_logistic)
)

print(
    "F1 Score:",
    f1_score(y_test, y_pred_logistic)
)

print(
    "ROC-AUC:",
    roc_auc_score(y_test, y_prob_logistic)
)

print(
    "PR-AUC:",
    average_precision_score(y_test, y_prob_logistic)
)
```

```python
print(
    classification_report(
        y_test,
        y_pred_logistic
    )
)
```

```python
# Create confusion matrix
cm = confusion_matrix(
    y_test,
    y_pred_logistic
)

# Display as a DataFrame
cm_df = pd.DataFrame(
    cm,
    index=["Actual Normal", "Actual Fraud"],
    columns=["Predicted Normal", "Predicted Fraud"]
)

cm_df
```

```python
model_df["risk_score"] = (
    model_df["fraud_probability"] * 100
)
```

```python
def assign_risk_level(score):

    if score >= 80:
        return "CRITICAL"

    elif score >= 60:
        return "HIGH"

    elif score >= 30:
        return "MEDIUM"

    else:
        return "LOW"
```

```python
model_df["risk_level"] = (
    model_df["risk_score"]
    .apply(assign_risk_level)
)
```

```python
def assign_priority(level):

    if level == "CRITICAL":
        return "URGENT INSPECTION"

    elif level == "HIGH":
        return "HIGH PRIORITY"

    elif level == "MEDIUM":
        return "MONITOR"

    else:
        return "LOW PRIORITY"
```

```python
model_df["inspection_priority"] = (
    model_df["risk_level"]
    .apply(assign_priority)
)
```
